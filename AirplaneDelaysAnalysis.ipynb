```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\since\Downloads\airlines_delays.csv")
```

```python
# Group the data by Airline and Status to sum the number of flights across cities
grouped_data = df.groupby(['Airline', 'Status']).sum()

# Display the grouped data
grouped_data
```

```python
# Create a bar plot comparing on-time and delayed flights
grouped_data.unstack().plot(kind='bar', figsize=(10, 6))

# Add title and labels to the plot
plt.title('On-Time vs Delayed Flights for Airlines Across Cities')
plt.ylabel('Number of Flights')
plt.xlabel('Airline and Status')
plt.xticks(rotation=0)
plt.tight_layout()

# Show the plot
plt.show()
```
